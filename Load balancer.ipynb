```python
gh = data['IPv4'].iloc[1] = data['IPv4 - Vserver hits'].loc[1]-data['IPv4 - Vserver hits'].loc[0]
gh = data['IPv4'].iloc[2] = data['IPv4 - Vserver hits'].loc[2]-data['IPv4 - Vserver hits'].loc[1]
data['IPv4'].loc[0] = 0
for i in range(1,718):
    data['IPv4'].iloc[i] = data['IPv4 - Vserver hits'].iloc[i]-data['IPv4 - Vserver hits'].iloc[i-1]
data['IPv4']
```

```python
data['IPv6'].loc[0]=0
for i in range(1,718):
    data['IPv6'].iloc[i] = data['IPv6 - Vserver hits'].iloc[i]-data['IPv6 - Vserver hits'].iloc[i-1]
data['IPv6']
```
